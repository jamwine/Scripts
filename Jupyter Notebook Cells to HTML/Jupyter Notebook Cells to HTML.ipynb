````python
import json
import re
import markdown2
import pandas as pd
````

## Converting Jupyter Cell to HTML

````python
def get_html(content):
    # Case: content is ```
    if len(content) < 2:
        mkdoc = '<hr />\n'
        return
    # Case: Ignoring the anchor tag - content[0]
    elif content[1].startswith('#'):
        mkdoc = ''.join(content[1:])
        question_number = re.findall(r'\d+', content[1])
    else:
        mkdoc = ''.join(content)
        question_number = re.findall(r'\d+', content[0])

    html = markdown2.markdown(mkdoc, extras=['fenced-code-blocks', 'markdown-in-html'])
    html = html.replace('InterviewImages/','https://raw.githubusercontent.com/jamwine/Data-Structures-and-Algorithm/master/InterviewImages/')
    if not question_number:
        question_number = f'Q{k}'
    else:
        question_number = question_number[0]
    with open(f'LC_{question_number}.html', "w", encoding='utf-8') as file:
        file.write(html)
    return html
````

````python
with open('Interview Questions.ipynb','r', encoding="utf8") as file:
    data = file.read()
    
cells = json.loads(data)['cells']
k = 1 # variable for extra_questions

for i in range(len(cells)):
    cell = cells[i]
    content = cell['source']
    try:
        get_html(content)
    except:
        print(i)
````

### Fetching Question Numbers to save it in CSV

````python
items = json.loads(data)['cells'][0]['source'][1:]
items
````

````python
df = pd.DataFrame(re.findall(r'\d+\.', ' '.join(items)), columns = ["Question"])
df["Question"] = df["Question"].apply(lambda x:x.strip('.'))
df
````
